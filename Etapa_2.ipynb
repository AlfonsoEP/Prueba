```python
# Convertir columnas a numérico limpiando los datos
def limpiar_a_float(serie):
    return pd.to_numeric(serie.astype(str).str.replace(r'[^\d\.-]', '', regex=True), errors='coerce')

# Columnas que necesitas limpiar y convertir a float
columnas_a_convertir = [
    'Cantidad Mercancía', 'KG C02', 'Precio Unitario', 'Rendimiento'
]

for col in columnas_a_convertir:
    if col in df_21_22.columns:
        df_21_22[col] = limpiar_a_float(df_21_22[col])

# Ahora actualizamos la lista de variables numéricas
numericas.extend([
    'Cantidad Mercancía', 'KG C02', 'Precio Unitario', 'Rendimiento'
])

# Asegúrate que estén presentes y en formato correcto
numericas = [col for col in numericas if col in df_21_22.columns and pd.api.types.is_numeric_dtype(df_21_22[col])]

# Estadísticas descriptivas actualizadas
print("🔹 Estadísticas descriptivas - Variables numéricas (actualizadas):\n")
print(df_21_22[numericas].describe().T[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']])
```

```python
# 4. VISUALIZACIONES

## 4.1 Boxplots (Variables numéricas)
for col in numericas:
    datos = df_21_22[col].dropna()
    if datos.nunique() > 1:
        plt.figure(figsize=(6, 2.5))
        sns.boxplot(x=datos)
        plt.title(f'Boxplot: {col}')
        plt.tight_layout()
        plt.show()

## 4.2 Histogramas (Variables numéricas)
for col in numericas:
    datos = df_21_22[col].dropna()
    if datos.nunique() > 1:
        plt.figure(figsize=(6, 3))
        sns.histplot(datos, kde=True, bins=30)
        plt.title(f'Histograma: {col}')
        plt.tight_layout()
        plt.show()

## 4.3 Diagramas de barras (Variables categóricas)
for col in categoricas:
    if df_21_22[col].nunique() < 30:  # para que no explote la gráfica
        plt.figure(figsize=(8, 4))
        df_21_22[col].value_counts().plot(kind='bar')
        plt.title(f'Distribución: {col}')
        plt.ylabel("Frecuencia")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```
